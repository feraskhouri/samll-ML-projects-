```python
# Import necessary libraries 
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
from sklearn.cluster import KMeans 
from sklearn.datasets import make_blobs 
```

Clustering is an unsupervised machine learning technique designed to group unlabeled examples based on their similarity to each other.

```python
X, y = make_blobs(n_samples=300, centers=4, random_state=42)
```

```python
print(X,y)
```

```python
print(type(X))
```

```python
# Visualize the dataset 
plt.scatter(X[:, 0], X[:, 1], s=50) 
plt.title("Generated Data for K-Means Clustering") 
plt.show() 
```

```python
# Implement K-Means clustering 
kmeans = KMeans(n_clusters=4) 
kmeans.fit(X) 
```

```python
# Get cluster centers and labels 
centers = kmeans.cluster_centers_ 
labels = kmeans.labels_ 
 
# Visualize the clustered data 
plt.scatter(X[:, 0], X[:, 1], c=labels, s=50, cmap='viridis') 
plt.scatter(centers[:, 0], centers[:, 1], c='red', marker='X', s=200, alpha=0.75) 
plt.title("K-Means Clustering Result") 
plt.show()
```

```python
# Import necessary library for hierarchical clustering
from sklearn.cluster import AgglomerativeClustering

# Implement Hierarchical Clustering
hierarchical = AgglomerativeClustering(n_clusters=4)
labels_hierarchical = hierarchical.fit_predict(X)

# Visualize the clustered data
plt.scatter(X[:, 0], X[:, 1], c=labels_hierarchical, s=50, cmap='viridis')
plt.title("Hierarchical Clustering Result")
plt.show()
```

```python
# Import necessary library for DBSCAN
from sklearn.cluster import DBSCAN

# Implement DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5)
labels_dbscan = dbscan.fit_predict(X)

# Visualize the clustered data
plt.scatter(X[:, 0], X[:, 1], c=labels_dbscan, s=50, cmap='viridis')
plt.title("DBSCAN Clustering Result")
plt.show()
```

Assignment 1: Algorithm Implementation and Visualization

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
```

```python
# Generate a sample dataset
X, _ = make_blobs(n_samples=300, centers=4, random_state=42, cluster_std=1.0)

# Visualize the dataset
plt.scatter(X[:, 0], X[:, 1], s=50)
plt.title("Original Dataset")
plt.show()
```

```python
# Implement K-Means Clustering
kmeans = KMeans(n_clusters=4, random_state=42)
kmeans.fit(X)

# Get cluster labels and centers
labels_kmeans = kmeans.labels_
centers_kmeans = kmeans.cluster_centers_

# Visualize K-Means Clustering Result
plt.scatter(X[:, 0], X[:, 1], c=labels_kmeans, s=50, cmap='viridis')
plt.scatter(centers_kmeans[:, 0], centers_kmeans[:, 1], c='red', marker='X', s=200, alpha=0.75)
plt.title("K-Means Clustering Result")
plt.show()
```

```python
# Implement Hierarchical Clustering
hierarchical = AgglomerativeClustering(n_clusters=4)
labels_hierarchical = hierarchical.fit_predict(X)

# Visualize Hierarchical Clustering Result
plt.scatter(X[:, 0], X[:, 1], c=labels_hierarchical, s=50, cmap='viridis')
plt.title("Hierarchical Clustering Result")
plt.show()
```

```python
# Implement DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5)
labels_dbscan = dbscan.fit_predict(X)

# Visualize DBSCAN Clustering Result
plt.scatter(X[:, 0], X[:, 1], c=labels_dbscan, s=50, cmap='viridis')
plt.title("DBSCAN Clustering Result")
plt.show()
```

    K-Means is ideal for well-separated clusters with uniform densities but fails with noise or irregular shapes.
    Hierarchical Clustering offers flexibility in understanding data structure but is less scalable.
    DBSCAN excels in handling noise and non-spherical clusters, making it suitable for complex datasets.

Visualizations provide an intuitive comparison of how each algorithm segments the data, highlighting their respective strengths and weaknesses.
